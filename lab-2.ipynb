withColumn and withoutColumn

```python
df1 = spark.read.csv("/Volumes/workspace/default/apacha-data/MegaMart.csv", header=True, inferSchema=True)
```

```python
display(df1)
```

```python
df1.withColumn("price_per_unit_new",df1.price_per_unit*2).show()
```

```python
display(df1.withColumn("price_per_unit_new",df1.price_per_unit*2))
```

```python
from pyspark.sql.functions import *
```

```python
df1.withColumn("country",lit("Pakistan")).show()
```

```python
df1.withColumn("increment",lit(1)).show()
```

```python
df1.printSchema()
```

```python
df2 = df1.withColumn("product_id",col("product_id").cast("integer"))
```

```python
df2.printSchema() 
```

```python
df1.withColumnRenamed("order_id","order_id_new").show()
```
