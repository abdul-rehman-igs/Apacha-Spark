# **Filter**

```python
df3 = spark.read.csv("/Volumes/workspace/default/apacha-data/MegaMart.csv", header=True, inferSchema=True)
```

```python
df3.show()
```

```python
display(df3.filter(df3.order_status == "Delivered"))
```

```python
display(df3.filter(df3.order_status != "Delivered"))
```

```python
display(df3.filter((df3.product_category == "Fashion")& (df3.payment_method == "Credit Card")))
```

```python
display(df3.filter((df3.product_category == "Fashion") | (df3.payment_method == "Credit Card")))
```

```python
display(df3.filter(df3.product_name.startswith("T")))
```

```python
display(df3.filter(df3.product_name.like("%es%")))
```

# **GroupBy**

```python
df8 = spark.read.csv("/Volumes/workspace/default/apacha-data/MegaMart.csv", header=True, inferSchema=True)
```

```python
df8.show()
```

```python
df8.groupBy("product_category").sum("quantity","price_per_unit").show()
```

```python
display(df8.groupBy("product_category","product_name").count())
```

```python
df8.groupBy("product_category").max("quantity","price_per_unit").show()
```

```python
df8.groupBy("product_category").min("quantity","price_per_unit").show()
```

```python
df8.groupBy("product_category").mean("quantity","price_per_unit").show()
```
